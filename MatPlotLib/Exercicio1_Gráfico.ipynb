```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Gas_Price_Mat/gas_prices.csv')
gas.head(5)
```

```python
plt.title('Price of Gas')

plt.plot(gas['Year'], gas['France'],'.--' ,label = 'FRA')
plt.plot(gas['Year'], gas['UK'], label = 'UK')
plt.plot(gas['Year'], gas['USA'], '.--', label = 'USA')
plt.plot(gas['Year'], gas['Japan'],'.-' ,label = 'JAP')
plt.plot(gas['Year'], gas['Mexico'], label = 'MEX')

plt.xticks(gas['Year'][::2])

plt.xlabel('Years')
plt.ylabel('Prices')

plt.legend()
plt.savefig('PriceG.pdf')
plt.show()
```
