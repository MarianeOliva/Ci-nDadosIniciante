```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Gas_Price_Mat/gas_prices.csv')
gas.head(5)
```

```python
plt.title('Price of Gas')

list_country = ['France', 'Japan', 'Mexico', 'UK', 'USA']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label=[country], marker ='.')

plt.xticks(gas['Year'][::2])

plt.xlabel('Years')
plt.ylabel('Prices')

plt.legend()
plt.savefig('Price_G.png')
plt.show()
```
