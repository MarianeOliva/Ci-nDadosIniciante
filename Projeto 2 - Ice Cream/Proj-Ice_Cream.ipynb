```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Aqui estou abrindo o dataset através do Pandas

df = pd.read_csv('IceCream/IceCreamData.csv')
```

```python
#celcius = (df['Revenue'] - 32) * 5/9
#celcius
```

```python
#Aqui estou filtrando dados em que a Temperatura é menor que 100, para diminuir a quantidade

ice = df.loc[(df['Temperature'] <= 100)]

print(ice)
```

```python
#Aqui realizo outro filtro, puxando apenas os primeiro 10 dados da lista

ice_cream = ice.head(10)

print(ice_cream)
```

```python
#Aqui é onde realizo o Gráfico

plt.title('Sales of Ice Cream') #Adicionando titulo.

plt.plot(ice_cream['Temperature'],  'r.--' ,label = 'US$') 

plt.xticks(ice_cream['Temperature'][::4])

plt.xlabel('Temperature (Fahrenheit)')
plt.ylabel('Sales (US$)')

plt.legend()
plt.savefig('IceSale.pdf')
plt.show()
```
